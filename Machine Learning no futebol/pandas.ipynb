```python
# pandas facilita na visualização dos dados
# usando tabelas e etc economizando muitas linhas de código
# se comparado ao fazer no excel

import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup as bs
from selenium import webdriver
import time

gecko = '/home/andre/Downloads/geckodriver'
driver = webdriver.FirefoxOptions()
driver.add_argument('-headless')
driver = webdriver.Firefox(executable_path=gecko, options=driver)

url = 'https://www.transfermarkt.pt/jumplist/startseite/wettbewerb/BRA1/'
driver.get(url)
time.sleep(5)
html = driver.execute_script('return document.documentElement.outerHTML')

soup = bs(html, 'html.parser')
tabela = pd.read_html(str(soup), header = 0)
tabela[3]
```
