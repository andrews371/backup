```python
# verificando o lucro e o prejuízo

def gerar_resultado(val_apostado, odds, resultado):
    '''retorna o valor após a aposta quanto de lucro ou prejuízo'''
    if resultado == 'Sim':
        lucro_preju = val_apostado * (odds - 1)
    else:
        lucro_preju = val_apostado * (-1)
    return lucro_preju
```

```python
fellabot['Resultado'] = fellabot.apply(lambda row: gerar_resultado(row['Valor da Entrada'], row['Odd'], row['Ganhou a aposta?']), axis = 1)
```
